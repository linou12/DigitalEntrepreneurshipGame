```python
import pandas as pd
import json

# Load the Excel file into a DataFrame
df = pd.read_excel('/Users/bina/Documents/DigitalEntrepreneurshipGame/src/ideas_and_mvps.xlsx')

# Rename the column
df.rename(columns={'Description or reason': 'Description'}, inplace=True)

# Drop unwanted columns, including 'ID' if it exists
columns_to_drop = ['Human Feedback', 'ID']  # Add 'ID' to the list of columns to drop
df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)

# Convert the DataFrame to the desired JSON format
formatted_data = {str(index + 1): row.to_dict() for index, row in df.iterrows()}

# Save the formatted JSON to a file (optional)
with open('/Users/bina/Documents/DigitalEntrepreneurshipGame/src/ideas_and_mvps_correct.json', 'w') as json_file:
    json.dump(formatted_data, json_file, indent=4)

# Print the formatted JSON (optional)
print(json.dumps(formatted_data, indent=4))
```
